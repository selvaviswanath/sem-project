```python
# -*- coding: utf-8 -*-
"""Phishing_Website.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1oKO03rCcbyeDKDGhek3rkjfJcr4zpXrz
"""
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
import pickle
df = pd.read_csv('https://raw.githubusercontent.com/IBM-EPBL/IBM-Project-1571-1658398856/main/Dataset%20Collection/dataset.csv')

df=df.iloc[:,[1,2,3,4,5,6,12,20,21,22,23,24,25,30,31]]

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score   
X=df.iloc[:,:13]
Y=df.iloc[:,13]

train_X,test_X,train_Y,test_Y=train_test_split(X,Y,test_size=0.40,random_state=10)

from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier()
model_4=rfc.fit(train_X,train_Y)

rfc_predict=model_4.predict(test_X)

print('The accuracy of Random Forest Classifier is: ' , 100.0 * accuracy_score(rfc_predict,test_Y))

print("Done by:\n")
print("Team id: PNT2022TMID26273\n")
print("Selva viswanath")
pickle.dump(rfc, open('model.pkl','wb'))
```
